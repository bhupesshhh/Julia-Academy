# Plotting

### Basics

````julia
#= 
Different ways to plot in Julia

with PyPlot
or else
=#
````

````julia
# We'll use Plots.jl module here
````

````julia
using Pkg
Pkg.add("Plots")
````

````julia
# Load Plots module
using Plots
````

````julia
x = -5:0.2:5
# Defining an independent array x
f(x) = x^2
# Defining a function
y = f.(x);
# Substituting the values to another independent array x
````

````julia
# Load the gr() backend
gr()
````

````julia
# plotting the graph
plot(x,y,lable="line")
scatter!(x,y,label="points")
````

````julia
#=
scatter! is mutating the existing plot
=#
````

### Plot with Random Data

````julia
temperature = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8];
````

````julia
pirates = [45000, 20000, 15000, 5000, 400, 17];
````

````julia
plot(pirates,temperature, label="line")
scatter!(pirates,temperature, lable="points")
````

````julia
xlabel!("Number of Pirates [Approximate]")
ylabel!("Global Temperature (C)")
title!("Influence of pirate population on global warming")
````

````julia
xflip!()
````
